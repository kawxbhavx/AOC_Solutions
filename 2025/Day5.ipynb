```python
#Part1
import pandas as pd
input = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""

fresh=[5,11,17]
spoiled=[1,8]
expected=len(fresh)
# print(expected)

with open("input_day5.txt", 'r', encoding='utf-8') as file:
  input = file.read()[:-1]

inputs = input.split("\n\n")
fresh_ranges=inputs[0].split("\n")
available=inputs[1].split("\n")

# print(fresh_ranges)
# print(available)
range_starts=[]
range_ends=[]
for fresh_range in fresh_ranges:
  range_start, range_end = fresh_range.split("-")
  range_starts.append(int(range_start))
  range_ends.append(int(range_end))
# print(range_starts)
# print(range_ends)

ranges=pd.DataFrame({"range_start":range_starts,"range_end":range_ends})
# print(ranges)
ranges_sorted=ranges.sort_values(by=["range_start","range_end"], ignore_index=True)
# print(ranges_sorted)

# for index,row in ranges_sorted.iterrows():
#   #print(row)
#   print(str(row["range_start"]) + "-" + str(row["range_end"]))

ranges_sorted["overlap_start"]=ranges_sorted["range_start"].copy()
ranges_sorted["overlap_end"]=ranges_sorted["range_end"].copy()

# print(ranges_sorted.shape[0])
for index in range(1,ranges_sorted.shape[0]):
  if ranges_sorted.at[index-1, "range_end"] +1 >= ranges_sorted.at[index, "range_start"]:
    ranges_sorted.at[index, "overlap_start"] = ranges_sorted.at[index-1, "overlap_start"]

for index in reversed(range(ranges_sorted.shape[0]-1)):
  if ranges_sorted.at[index, "overlap_start"] == ranges_sorted.at[index+1, "overlap_start"]:
    if ranges_sorted.at[index, "overlap_end"] > ranges_sorted.at[index+1, "overlap_end"]:
      ranges_sorted.at[index+1, "overlap_end"] = ranges_sorted.at[index, "overlap_end"]
    else:
      ranges_sorted.at[index, "overlap_end"] = ranges_sorted.at[index+1, "overlap_end"]

# for index,row in ranges_sorted.iterrows():
#   print(f'{row["range_start"]}-{row["range_end"]}-{row["overlap_start"]}-{row["overlap_end"]}')

ranges_unique=ranges_sorted.drop_duplicates(["overlap_start","overlap_end"],ignore_index=True)
# for index,row in ranges_unique.iterrows():
#   print(str(row["range_start"]) + "-" + str(row["range_end"]))

available = [int(a) for a in available]
# print(len(available))
fresh = []
spoiled = []
fresh_count=0
for available_item in available:
  for index,row in ranges_unique.iterrows():
    if available_item >= (row["overlap_start"] -1) and available_item <= (row["overlap_end"] +1):
      fresh_count+=1
      fresh.append(available_item)
      break
spoiled = [a for a in available if a not in fresh]
# print(fresh)
# print(spoiled)
# print(fresh_count)
print(len(fresh))

#Part2
all_count = 0
for index,row in ranges_unique.iterrows():
  all_count += row["overlap_end"] - row["overlap_start"] + 1
print(all_count)
```
