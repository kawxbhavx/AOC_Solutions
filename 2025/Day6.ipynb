```python
#Part1
input = """123 328  51 64
 45 64  387 23
  6 98  215 314
*   +   *   +  """

expected=[33210, 490, 4243455, 401]
# print(sum(expected))

with open("input_day6.txt", 'r', encoding='utf-8') as file:
  input = file.read()[:-1]

inputs = input.split('\n')
operators = inputs[-1]
operators = operators.split()
# print(operators)
vals = []
inputs = inputs[:-1]
for row in inputs:
  elements = row.split()
  vals.append([int(e) for e in elements])

res = [1 if o=="*" else 0 for o in operators]
# print(res)
for row in vals:
  for i in range(len(operators)):
    if operators[i]=="*":
      res[i] *= row[i]
    else:
      res[i] += row[i]
print(sum(res))
```

```python
#Part2
input = """123 328  51 64
 45 64  387 23
  6 98  215 314
*   +   *   +  """

expected=[8544, 625, 3253600, 1058]
# print(sum(expected))

with open("input_day6.txt", 'r', encoding='utf-8') as file:
  input = file.read()[:-1]

inputs = input.split('\n')
operators = inputs[-1]
inputs = inputs[:-1]

transposed=[]
# transposed=(' ' * len(operators)).split(" ")
for c in range(len(operators)):
  num = ""
  for r in range(len(inputs)):
    # print(r,c)
    num = num + inputs[r][c]
  transposed.append(num)
# for val in transposed:
#   print(val)

res = []
i=0
while i < len(operators)-1:
  if operators[i]=="*":
    res_item = 1
    if len(transposed[i].replace(" ",""))>0:
      res_item = int(transposed[i])
    i += 1
    try:
      while operators[i]==" ":
        if len(transposed[i].replace(" ",""))>0:
          res_item *= int(transposed[i])
        i += 1
      res.append(res_item)
    except:
      res.append(res_item)
  else:
    res_item = 0
    if len(transposed[i].replace(" ",""))>0:
      res_item = int(transposed[i])
    i += 1
    try:
      while operators[i]==" ":
        if len(transposed[i].replace(" ",""))>0:
          res_item += int(transposed[i])
        i += 1
      res.append(res_item)
    except:
      res.append(res_item)
print(sum(res))
```
