```python
#Part1

start=50
input="""L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""
count=0

with open("input.txt", 'r', encoding='utf-8') as file:
  input = file.read()[:-1]
seq=input.split("\n")
for steps in seq:
  remaining_steps=int(steps[1:])
  if steps[0]=="R":
    while remaining_steps>0:
      start+=1
      if start==100:
        start=0
      remaining_steps-=1
    #print(start)
    if start==0:
      count+=1
  elif steps[0]=="L":
    while remaining_steps>0:
      start-=1
      if start==-1:
        start=99
      remaining_steps-=1
    #print(start)
    if start==0:
      count+=1
print(count)
```
