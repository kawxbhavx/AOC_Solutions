```python
#Part1
input = """987654321111111
811111111111119
234234234234278
818181911112111"""
expected_arr=[98,89,78,92]
expected=sum(expected_arr)
expected

with open("input.txt", 'r', encoding='utf-8') as file:
  input = file.read()[:-1]

grid_rows=input.split("\n")
max_joltages=[]
for grid_row in grid_rows:
  joltages=[]
  for i in range(len(grid_row)-1):
    for j in range(i+1,len(grid_row)):
      joltages.append(int(grid_row[i:i+1]+grid_row[j:j+1]))
  max_joltages.append(max(joltages))
print(sum(max_joltages))
```

```python
#Part2.1
input = """987654321111111
811111111111119
234234234234278
818181911112111"""
expected_arr=[987654321111,811111111119,434234234278,888911112111]
expected=sum(expected_arr)
print(expected)

#with open("input.txt", 'r', encoding='utf-8') as file:
  #input = file.read()[:-1]

grid_rows=input.split("\n")
max_joltages=[]
for grid_row in grid_rows:
  eliminated=[]
  for i in range(len(grid_row)):
    eliminated.append(grid_row[:i]+grid_row[i+1:])
    #print(eliminated[-1])
  eliminated=[e for e in eliminated if len(e) > 12]
  while len(eliminated)>0:
    next_eliminated=[]
    for remaining in eliminated:
      for i in range(len(remaining)):
        next_eliminated.append(remaining[:i]+remaining[i+1:])
        #print(next_eliminated[-1])
    eliminated=[e for e in next_eliminated if len(e) > 12]
  max_joltages.append(max([int(e) for e in next_eliminated]))
print(sum(max_joltages))
```
